```python
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("../input/trainset/trainset.csv")
train.head(10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train.iloc[:,1:],train.iloc[:,0], test_size=0.1, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_train = X_train.values.reshape(len(X_train),28,28,1)
X_test = X_test.values.reshape(len(X_test),28,28,1)
```

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
```

```python
model.summary()
```

```python
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=10, 
                    validation_data=(X_test, y_test))
```

```python
test=pd.read_csv("../input/testset/testset.csv")
test = test.values.reshape(len(test),28,28,1)
```

```python
prediction = model.predict_classes(test)
```

```python
prediction
```

```python
result=pd.DataFrame({"ImageId":range(1,len(prediction)+1),"Label":prediction})
```
